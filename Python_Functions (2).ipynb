5.  What are iterators in Python and how do they differ from iterables?

 7.  What are the advantages of using generators over regular functions?

 =>Generators offer several advantages over regular functions, primarily concerning memory efficiency and lazy evaluation. Unlike regular functions that return a complete result immediately, generators yield values one at a time, saving memory, especially when dealing with large datasets or infinite sequences. They also allow for lazy evaluation, computing values only when needed, potentially improving performance.

  * Memory Efficiency:
Regular functions create a complete result in memory before returning it, which can be problematic for large datasets that consume significant memory.
  
  * Improved Readability and Conciseness:
Generators can sometimes lead to more concise and readable code compared to using explicit loops or iterators.

 10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

 =>map(), filter(), and reduce() are built-in functions in Python that operate on iterables (like lists, tuples, etc.) by applying a function to the items within. Each has a distinct purpose:

  map():

   Applies a given function to each item in an iterable and returns an iterator that yields the results. It transforms each element individually.

  filter():

  Constructs an iterator from elements of an iterable for which a function returns true. It effectively selects elements based on a condition.

  reduce():

  Applies a function of two arguments cumulatively to the items of an iterable, from left to right, to reduce the iterable to a single value. It performs an aggregation.

  * example:-

      from functools import reduce

       numbers = [1, 2, 3, 4, 5]

        even_numbers = list(filter(lambda x: x % 2 == 0, numbers)) # Result: [2, 4]

       reduce(): calculates the product of all numbers

```python
# 4. Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(n):
  if n >= 1 and n <= 200:
    if n <= 1:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

is_prime(7)



```

```python
  # 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
class Fibonacci:
    def __init__(self, limit):
        self.limit = limit  # How many terms to generate
        self.count = 0      # Counter to track how many terms we've generated
        self.a = 0          # First Fibonacci number
        self.b = 1          # Second Fibonacci number

    def __iter__(self):
        return self  # An iterator must return itself

    def __next__(self):
        if self.count >= self.limit:
            raise StopIteration  # Stop when we reach the limit
        if self.count == 0:
            self.count += 1
            return 0
        elif self.count == 1:
            self.count += 1
            return 1
        else:
            self.count += 1
            next_value = self.a + self.b
            self.a = self.b
            self.b = next_value
            return next_value

# Example: Print first 10 Fibonacci numbers
fib = Fibonacci(12)
for number in fib:
    print(number)


```

```python
#6.Write a generator function in Python that yields the powers of 2 up to a given exponent.

def powers_of_two(given_exponent):
    for i in range(given_exponent + 1):
        yield 2 ** i

for power in powers_of_two(10):
    print(power)
```

```python
#7.Implement a generator function that reads a file line by line and yields each line as a string.

def read_file_line_by_line(filename):
    with open(filename, 'r') as file:
        for line in file:
            yield line.strip()
```

```python
# 10.Create a Python program that uses `filter()` to remove all the vowels from a given string.

def remove_vowels(input_str):
    vowels = "aeiouAEIOU"
    return ''.join(filter(lambda char: char not in vowels, input_str)) # here i have used the filter function

text = "Hello, World!"
no_vowels = remove_vowels(text)
print("Without vowels:", no_vowels)
```
